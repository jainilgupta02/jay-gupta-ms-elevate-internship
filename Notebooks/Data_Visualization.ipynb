```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
#Line Plot
plt.plot(data['total_bill'])
plt.title("Total Bill Trend")
plt.xlabel("Customer index")
plt.ylabel("Bill Amount")
plt.show()
```

```python
#Scatter Chart
plt.scatter(data['total_bill'],data['tip'])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#Bar chart
data['day'].value_counts().plot(kind='bar')
plt.title("Customers per day")
plt.xlabel("Day")
plt.ylabel("Number of Customers")
plt.show()
```

```python
#use cases: sales per day, orders per category, users from which region
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(10,5))
plt.plot(data['total_bill'], color = 'green', marker = 'o', linestyle = '--')
plt.title("Styled line plot")
plt.xlabel("Index")
plt.ylabel("Total Bill")
plt.grid(True)
```

```python
#Histogram distribution
plt.hist(data['total_bill'], bins = 20)
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

#Seaborn

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data)
plt.title("Boxplot of Total Bill by Day")
```

```python
sns.countplot(x = 'day', data = data)
plt.title("customer count by day")
```

```python
#Voilin plot
sns.violinplot(x = 'sex', y = 'tip', data = data)
plt.title("Violin plot of tip by sex")
```

```python
corr = data.corr(numeric_only=True)
```

```python
#Heat plot
sns.heatmap(corr, annot = True, cmap = "coolwarm")
plt.title("Correlation Heatmap")
```

Do smokers tip more

```python
sns.boxplot(x = 'smoker', y = 'tip', data = data)
plt.title("Boxplot of Tip by Smoker Status")
```

```python
sns.barplot(x = 'smoker', y = 'tip', data = data)
plt.title("Bar Chart of Tip by Smoker Status")
```

Gender wise spending

```python
sns.violinplot(x = 'sex', y = 'total_bill', data = data, color = 'blue')
plt.title("Violin Plot of Total Bill by Gender")
```

```python
sns.scatterplot(x='sex', y='total_bill', hue='smoker', data = data)
plt.title("Scatter Plot of Total Bill by Gender and Smoker Status")
```
